## **Section 1: Function Using If/Else**

In this lab, you will create a Python program that simulates a Magic 8-Ball, where the secret logic behind the "random" response is determined based on the length of the user's question. This exercise will help you practice defining functions and using conditionals.

### **Instructions:**
1. Write a Python function called `magic_8_ball` that takes one parameter:
    - `question`: a string representing the user's question.

2. Inside the function, determine the length of the user's question using the len() function, and store this value in a variable.

3. Use conditional statements to select a response based on the length of the question:

    - If the question is shorter than 10 characters, the response should be "Signs point to yes"
    - If the question is 10 characters or longer but shorter than 20 characters, the response should be "Ask again later"
    - If the question is 20 characters or longer, the response should be "Don't count on it"

4. Return the selected response.

5. Back outside of the function, prompt the user to enter a yes-or-no question.

6. Call the `magic_8_ball` function with the user's question as the argument.

7. Display the Magic 8-Ball's response to the user.

8. **Optional Challenge:** Use recursion to ask the user "Would you like to ask another question?" after each answer, and keep going until they respond "No".
