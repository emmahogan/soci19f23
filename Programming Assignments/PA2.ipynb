# **Problem 1**: Daily Fitness

In this problem, you will write a Python program to assist someone in reaching a personal daily fitness goal. Your program will do the following:


> A.   Gather information from the user on their personal fitness goal for the day, their preferred form of exercise, and the duration they have exercised so far.
>
> B.   Compute how many calories the user has burned so far, and report to them how much progress they have made towards their goal.
>
> C.   Calculate how much longer the user needs to do this exercise in order to reach their daily goal.



## Part A: Gathering Information

Write some lines of code to begin your program by gathering information from the user regarding their personal fitness goal for the day, preferred form of exercise, and the duration they have exercised so far:

1. Prompt the user to enter their daily fitness goal in calories (an integer value representing their target calorie burn for the day).
2. Ask the user to input their preferred form of exercise (e.g., "Walking", "Running", "Push-ups").
3. Obtain the duration (in minutes) the user has exercised so far.

## Part B: Calories Burned Calculator

In this part, you will define a function that calculates the calories burned based on the exercise type and duration, utilizing a conditional statement inside the function:

1. Define a function called **calculate_calories_burned** that takes two parameters:
    - *exercise_type (string)*: representing the type of exercise (i.e., "Walking", "Running", or "Push-ups")
    - *duration_minutes (float)*: representing the duration of the exercise in minutes.
2. Utilize a conditional statement (if, elif, else) inside the function to determine the appropriate calorie burn rate based on the exercise type. Use the following rates:
    - Walking: 5.0 calories/minute
    - Running: 10.0 calories/minute
    - Push-ups: 7.0 calories/minute
3. Calculate the calories burned using the determined rate and the provided duration.
4. **Return** the calculated calories burned.

Note: For the purposes of this assignment, you can assume that the user has entered "Walking", "Running", or "Push-ups". As an optional challenge (not graded), you can add an additional case to your conditional statement that deals with the case that the exercise_type is not one of these three (for example, perhaps you would print a message to the user saying "Sorry, I don't know that exercise. Please choose either Walking, Running, or Push-ups").

## Part C: Report Progress

In this part, you will calculate the progress towards the daily fitness goal as a percentage and provide feedback to the user:

1. Call the calculate_calories_burned function (defined in Part 2) with appropriate arguments for exercise type and duration.
2. Calculate the progress towards the daily fitness goal as a percentage using the following formula:

$\text{progress_percentage} = \left( \frac{\text{calories_burned_so_far}}{\text{daily_goal}} \right) \times 100$

3. Display a message to the user indicating their progress towards the goal as a percentage. The message should be in the following format (but feel free to personalize it as you wish):



```
You are 68% of the way towards your daily fitness goal, keep going!
```


## Optional Challenges

As an optional challenge (not graded), consider the following possible additions to your program:
- Print a different message to the user based on their amount of progress
- Determine whether the user has reached their goal, and print a different message accordingly
- Factor in more variables to your function in Part 2 (such as age, weight, etc.)
- If the user has not reached their goal yet, compute how much longer they have to spend doing their exercise in order to reach the goal
