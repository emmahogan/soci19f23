# **Problem 1**: Analyzing Real-World Data

**Objective:** In this assignment, you will delve into data analysis in Python, focusing on the distribution of incarceration rates in different countries. You will be given lists of country names and their corresponding incarceration rates (number of people incarcerated per 100,000 people). **Instructions:**

1. You will be provided with two lists: `countries` (containing country names) and `incarceration_rates` (containing the corresponding incarceration rates). There is **no need to re-write** these lines: assume that the following lines are at the top of the program:


```
countries = ["China", "Mexico", "Italy", "India", "Ghana", "Brazil", "South Africa",
              "Japan", "Iran", "Russia", "New Zealand", "United States"]

incarceration_rates = [121, 166, 89, 35, 42, 357, 248, 38, 228, 329, 188, 664]
```


2. **Define a function** called `compute_std` that takes one parameter, `data`, which is the list of values to compute the standard deviation of.

4. Next, **create an empty list** in a new variable called `distances`. This list will eventually hold the distance of each country's incarceration rate from the mean.

5. **Write a for loop** to iterate through each value in the list `data`. Inside the loop, calculate the distance of the value from the mean, and append it to the `distances` list.

- *Hint: Recall that the distance from the mean will be a positive value regardless of whether it is above or below the mean. Therefore, we can use the built-in **abs()** function to get the absolute value of the difference between each rate and the mean.*

6. After the loop, **return the standard deviation**. Since the standard deviation is the average distance from the mean, calculate this value by finding the average of the `distances` list.

7. **Call your function**, passing the `incarceration_rates` list as the argument for `data`, and store the result in a new variable called `std_rate`.

8. **Print the resulting standard deviation** with an appropriate message.
